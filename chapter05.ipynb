```python
df = pd.DataFrame({'var1' : [4,3,8],
                   'var2' : [2,6,11]})
df
```

```python
# 파생변수 만들기
df['var_sum'] = df['var1'] + df['var2']
df['var_mean'] = df['var1'] + df['var2'] / 2
df
```

```python
mpg['total'] = (mpg['cty'] + mpg['hwy']) / 2 # 통합연비변수 만들기
mpg.head()
```

```python
# 파생변수의 평균 구하기 (둘다 같은거)
sum(mpg['total']) / len(mpg)
mpg['total'].mean()
```

```python
mpg['total'].describe()
```

```python
mpg['total'].plot.hist()
```

```python
# total 이 20점 이상이면 pass, 아니면 fail 을 부여하는 파생변수 test 만들기
mpg['test'] = np.where(mpg['total'] >= 20, 'pass', 'fail')
mpg.head()
```

```python
# 빈도표 분석으로 test 합격 판정 개수 확인
mpg['test'].value_counts()
```

```python
# 막대 그래프로 빈도 표현
count_test = mpg['test'].value_counts()
count_test.plot.bar()
```

```python
# 축 이름 수평으로 회전
count_test.plot.bar(rot = 0)
```

```python
# 중첩 조건문
# total이 30이상이면 A, 25이상이면 B, 20이상이면 C 아니면 D
mpg['grade2'] = np.where(mpg['total'] >= 30, 'A',
                np.where(mpg['total'] >= 25, 'B',
                np.where(mpg['total'] >= 20, 'C', 'D')))
mpg.head()
```

```python
# 목록에 해당하는 행으로 변수 만들기
# category가 compact거나 subcompact거나 2seater면 small, 아니면 large
mpg['size'] = np.where((mpg['category'] == 'compact')|
                       (mpg['category'] == 'subcompact')|
                       (mpg['category'] == '2seater'),
                       'small', 'large')
mpg['size'].value_counts()
```

```python
# df.isin() 함수로 목록에 해당하면 small, 아니면 large
mpg['size'] = np.where(mpg['category'].isin(['compact', 'subcompact', '2seater']),
                       'small', 'large')
mpg['size'].value_counts()
```
